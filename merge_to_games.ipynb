## Импорты

```python
import pandas as pd
import numpy as np
```

## Читаем датамсеты делаем маппинг

```python
games = pd.read_csv('data/games.csv')
games['first_release_date'] = pd.to_datetime(games['first_release_date'], unit='s')
games['game_modes'] = games['game_modes'].apply(lambda x: x.replace('[', '').replace(']', '') if pd.notna(x) else x)
games['genres'] = games['genres'].apply(lambda x: x.replace('[', '').replace(']', '') if pd.notna(x) else x)
games['platforms'] = games['platforms'].apply(lambda x: x.replace('[', '').replace(']', '') if pd.notna(x) else x)
games['themes'] = games['themes'].apply(lambda x: x.replace('[', '').replace(']', '') if pd.notna(x) else x)
games.head()
```

```python
game_modes = pd.read_csv('data/game_modes.csv')
game_modes_map = dict(zip(game_modes['id'].astype(str), game_modes['name']))
game_modes_map
```

```python
genres = pd.read_csv('data/genres.csv')
genres_map = dict(zip(genres['id'].astype(str), genres['name']))
genres_map
```

```python
platforms = pd.read_csv('data/platforms.csv')
platforms_map = dict(zip(platforms['id'].astype(str), platforms['name']))
platforms_map
```

```python
themes = pd.read_csv('data/themes.csv')
themes_map = dict(zip(themes['id'].astype(str), themes['name']))
themes_map
```

## Заменяем в основном датасете

Делаем функцию для замены

```python
def replace_ids(x, mapping):

    if pd.isna(x):
        return x
    
    col = x.split(', ')
    names = []
    
    for i in col:
        if i in mapping:
            names.append(mapping[i])
        else:
            names.append(i)
    res = ', '.join(names)
    return res
```

```python

games['game_modes'] = games['game_modes'].apply(replace_ids, mapping=game_modes_map)
games['genres'] = games['genres'].apply(replace_ids, mapping=genres_map)
games['platforms'] = games['platforms'].apply(replace_ids, mapping=platforms_map)
games['themes'] = games['themes'].apply(replace_ids, mapping=themes_map)
games.head()
```

```python
games.to_csv('data/games_complete.csv', index=False)
```
